```python
import pandas as pd
import numpy as np
```

```python
file1= pd.read_csv("file1.csv")
display(file1.head())
file1.shape
#tenemos que dividir
```

```python
file1['Customer Lifetime Value']=file1['Customer Lifetime Value'].str.replace("%","") 
```

```python
file1['Customer Lifetime Value'] =  pd.to_numeric(file1['Customer Lifetime Value'], errors='coerce')/100
file1.head()
```

```python
cols1 = []
for c in file1.columns:
    cols1.append(c.lower())
cols1
file1.columns = cols1
file1 = file1.rename (columns ={"st":"state"})
file1.columns
```

```python
file1 = file1[['customer', 'state', 'gender', 'education', 'customer lifetime value','income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
display(file1.head ())
```
